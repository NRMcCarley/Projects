```python
import numpy as np
import pandas as pd
import random
```

```python
sheet1_url = "https://docs.google.com/spreadsheets/d/1sqdLbPDdzUsxxAMHg0Ah8pYhTjnynz8mziTtbSEQBtY/export?format=csv&gid=0"

df1 = pd.read_csv(sheet1_url)[1:]

df1
```

```python
brother_list = []

for brother in df1['Name']:
  brother_list.append(brother)

brother_list
```

```python
sheet2_url = "https://docs.google.com/spreadsheets/d/1sqdLbPDdzUsxxAMHg0Ah8pYhTjnynz8mziTtbSEQBtY/export?format=csv&gid=1992876087"

df2 = pd.read_csv(sheet2_url)

df2
```

```python
job_list = []

for job in df2['Job']:
  job_list.append(job)

job_list
```

```python
selection_weights = df1['Cleaning Share'].to_numpy()

selection_weights = np.array([float(value.replace('%', '')) for value in selection_weights]) / 100

selection_weights /= sum(selection_weights)

selection_weights
```

```python
def randomBro():
  return random.choices(brother_list, weights=selection_weights)[0]
```

```python
randomBro()
```

```python
schedule = np.zeros((len(job_list), len(brother_list)))
numWeeks = 7

num_assignments = numWeeks * len(job_list)
```

```python
jobs_per_person = np.round(num_assignments * selection_weights)

num_to_change = int(num_assignments - sum(jobs_per_person))

if num_to_change < 0:
  for i in range(abs(num_to_change)):
    running = True
    while running:
      idx = np.random.choice(range(len(brother_list)), 1, p=selection_weights)[0]
      if jobs_per_person[idx] >= 2:
        jobs_per_person[idx] -= 1
        running = False
else:
  for i in range(abs(num_to_change)):
    idx = np.random.choice(range(len(brother_list)), 1, p=selection_weights)[0]
    jobs_per_person[idx] += 1
```

```python
schedule = -1*np.ones((numWeeks, len(job_list)), dtype=np.int32)     # (weeks, jobs)
```

```python
for bro in range(len(brother_list)):
  print(f'Brother: {bro}')
  num_jobs = int(jobs_per_person[bro])
  print(f'Jobs: {num_jobs}')

  # Determine open jobs (j) at which weeks (i)
  available_slots = np.argwhere(schedule == -1)
  num_available = len(available_slots)
  print(f'Number of available indices: {num_available}')

  # Check if there are enough available slots for the requested number of jobs
  if num_available < num_jobs:
      raise ValueError(f"Not enough available slots for brother {bro}")

  # Generate unique indices for jobs to be assigned to
  job_idxs = np.random.choice(num_available, size=num_jobs, replace=False)
  print(f'Number of jobs assigned to brother {bro}: {len(job_idxs)}')

  # Iterate through indices and assign the jobs to the brother
  for idx in job_idxs:
      i, j = available_slots[idx]
      schedule[i, j] = bro


schedule
```

```python
for week in range(numWeeks):
  print(f'WEEK {week+1} SCHEDULE:')
  for job_idx in range(len(job_list)):
    brother = schedule[week, job_idx]
    print(f'{job_list[job_idx]}: {brother_list[brother]}')
  print('\n')
```
